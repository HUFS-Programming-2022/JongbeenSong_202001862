## 여러가지 함수

- count(): 텍스트에서 어떤 요소의 개수를 세는 함수
- replace(): A -> B
- strip(): 불필요한 공백, 등등 원하는 요소 제거하는 역할
- find(): 원하는 문자열 찾기
- index(): 원하는 문자열의 인덱스 반환

## 함수 II

- concordance() 특정 스트림이 어떤 문맥에서 쓰이는지
- similar() 주어진 스트림과 비슷한 문맥에서 쓰이는 단어들 보여줌
- common_context(리스트) 주어진 스트림들이 공유하는 문맥
- dispersion_plot(리스트) 주어진 스트림들이 텍스트의 어느 위치에 나오는지 이미지로 보여줌

```python
# 함수 활용 - 개인 연습
info = 'My name is Jongbeen, and I live in Seoul     '
info_refined = info.rstrip()
info_replaced = info_refined.replace('Jongbeen', 'James')
nvm_of_i = info_replaced.count('i')
address = info_replaced.find('Seoul')
name = info_replaced.index('name')
```

```python
print(info_refined)
print(info_replaced)
print(nvm_of_i)
print(address)
print(name)
```

### 리스트 & 리스트 컴프리헨션

- list + for loop
- [A for B in C]
    - C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A: 내가 원하는 형태
    - B: C의 형태
    - C: 반복가능한 자료형 (리스트)
    
- 조건 추가: [A for A' in C if ...]

```python
# 리스트 컴프리헨션 개인연습
sorted(set([word.lower() for word in text1.tokens if 'ist' in word]))
```

###  FreqDist(test)

- Frequency Distribution: 단어가 몇번 출현했는지 (빈도수)
- 텍스트의 성질 알기 위해
    - 단어 개수
    - 단어 빈도수

```python
fdist = FreqDist(text4)
fdist.most_common(10)
```
