```python
import nltk
nltk.download('book', quiet=True)
```

```python
from nltk import book
```

```python
moby = book.text1
```

```python
moby.tokens # 단어에 접근
"""
토큰? 프로그래밍에서 정의하는 단어
    "I'm a boy." I'm, I am, I 'm
"""
```

```python
# 총 길이
len(moby.tokens)
```

```python
# 중복되는 단어들을 걸러내자
set(moby.tokens)
```

- set
    - {a, b, c, ...}
    - 중복 불가능
    - 수학의 집합과 동일한 개념
    - 순서 없음
- list
    - [a, b, c, ...]
    - 중복 가능
    - 순서 있음

```python
cars = ['kia', 'benz', 'kia', 'porsche']
set(cars)
```

```python
# 중복없는 단어 수
len(set(moby.tokens))
```

```python
# lexical diversity
len(set(moby.tokens)) / len(moby.tokens)
```

```python
# 오름차순, 내림차순 정렬
sorted(cars)
sorted(cars, reverse=True)
```

```python
# mobydick 역순
sorted(set(moby.tokens), reverse=True)[:10]
```
